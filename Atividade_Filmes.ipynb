```python
#Selecionando a base de dados
db = cliente.get_database('filmes')
```

```python
#Insere os documentos na nova base de dados
db.filmes.insert_many(dicionario)
```

```python
#Listando todos os filmes existentes
list(db.filmes.find())
```

```python
#Definindo a coleção "filmes" na variável collection
collection = db.get_collection('filmes')
```

```python
#Inserindo um filme novo na lista
add_filme = {
    'Título do Filme': 'Os Incríveis',
    'Ano de Lançamento': 2004,
    'Bilheteria (US$)': 631000000.0,
    'Duração (minutos)': 115,
    'Aprovação no Rotten Tomatoes': '97%'
}
result = collection.insert_one(add_filme)
```

```python
#Selecionado apenas os filmes com uma avaliação no Rotten Tomatoes maior que 90%
filmes_90 = collection.find({'Aprovação no Rotten Tomatoes': {'$gt': '90%'}})
for filme in filmes_90:
    print(filme)
```

```python
#Excluindo o filme 'O Círculo' da lista
collection.delete_one({"Titulo do Filme": "O Círculo"})
```

```python
#Atualizando a avaliação no Rotten Tomatoes do filme 'Wish: O Poder dos Desejos' para 80%
collection.update_one({'Título do Filme': 'Wish: O Poder dos Desejos'}, {'$set': {'Aprovação no Rotten Tomatoes': '80%'}})
```
